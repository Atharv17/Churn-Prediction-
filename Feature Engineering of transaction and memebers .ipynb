# Feature engineering using Transaction and Members Table 

```python
import pandas as pd 
import numpy as np 
import pickle 
```

```python
train = pd.read_csv('train.csv')
```

```python
transaction = pd.read_csv('transactions.csv')
```

```python
member = pd.read_csv('members_v3.csv')
```

### Extracting features from Transaction 

#### Transaction before Jan,2017

```python
transaction_before_jan17 = transaction[transaction['transaction_date'].le(20161231)]
```

```python
transaction_before_jan17
```

```python
## Discount Feature 

transaction_before_jan17['discount']= transaction_before_jan17['plan_list_price']-transaction_before_jan17['actual_amount_paid']
```

```python
a = transaction_before_jan17[transaction_before_jan17['discount']<0]
```

```python
a.head(-1)
```

#### 857421 rows has negative discount, which is not possible so removing those rows

```python
transaction_before_jan17 = transaction_before_jan17[transaction_before_jan17['discount']>-1]
```

```python
transaction_before_jan17.head(-2)
```

```python
## is_discount : To check weather the discount is given to customer or not 

transaction_before_jan17['is_discount'] = transaction_before_jan17.discount.apply(lambda x:1 if x>0 else 0)
transaction_before_jan17.head(-3)
```

```python
### Membership Duration of a customer 

transaction_before_jan17['membership_duration'] = transaction_before_jan17.membership_expire_date - transaction.transaction_date
transaction_before_jan17.head(-3)
```

```python
transaction_members = pd.merge(transaction_before_jan17,member,on='msno')
```

```python
del transaction_before_jan17
del member
```

```python
transaction_members.head(-2)
```

```python
## Customer who have auto-renewed and not cancelled 

transaction_members['auto-renew_not_cancel'] = ((transaction_members.is_auto_renew==1)==(transaction_members.is_cancel==0)).astype(np.int8)
```

```python
transaction_members.head(2)
```

```python
## Customer who have not auto_renew_cancel 

transaction_members['not_auto-renew_cancel'] = ((transaction_members.is_auto_renew==0)==(transaction_members.is_cancel==1)).astype(np.int8)
```

```python
transaction_members.head(-2)
```

```python
## Finding the counts of How many times a we have a particular customer imformation 
counts_transaction =transaction_members.groupby(['msno'], as_index=False)['actual_amount_paid','payment_plan_days','discount'].count()
 
```

```python
counts_transaction.head(2)
```

```python
## Finding the sum for finding the Average Later 
sum_transaction = transaction_members.groupby(['msno'],as_index=False)['actual_amount_paid','payment_plan_days','discount'].sum()
```

```python
sum_transaction.head(2)
```

```python
transaction_members[['actual_amount_paid_avg','payment_plan_days_avg','discount_avg']] = sum_transaction[['actual_amount_paid','payment_plan_days','discount']]/counts_transaction[['actual_amount_paid','payment_plan_days','discount']]
```

```python
transaction_members.head()
```

```python
transaction_members = pd.merge(transaction_members,counts_transaction,on='msno')
```

```python
transaction_members = pd.merge(transaction_members,sum_transaction,on='msno')
```
