```python
def featurization_transac(transaction,members):
    
    
    
    
    
    transaction['discount'] = transaction['plan_list_price']-transaction['actual_amount_paid']
    
    transaction = transaction[transaction['discount']>-1]
    
    transaction['is_discount'] = transaction.discount.apply(lambda x:1 if x>0 else 0)
    
    
    
    transaction['membership_duration'] =transaction.membership_expire_date-transaction.transaction_date
    
    transaction = pd.merge(transaction,members,on='msno')
    
    transaction['auto-renew_not_cancel'] = ((transaction.is_auto_renew==1)==(transaction.is_cancel==0)).astype(np.int8)
    
    transaction['not_auto-renew_cancel']=((transaction.is_auto_renew==0)==(transaction.is_cancel==1)).astype(np.int8)
    
    count_trans = transaction.groupby(['msno'],as_index=False)['actual_amount_paid','payment_plan_days','discount'].count()
    
    sum_trans = transaction.groupby(['msno'],as_index=False)['actual_amount_paid','payment_plan_days','discount'].sum()
    
    transaction[['actual_amount_paid_avg','payment_plan_days_avg','discount_avg']] = sum_trans[['actual_amount_paid','payment_plan_days','discount']]/count_trans[['actual_amount_paid','payment_plan_days','discount']]
    
    transaction = pd.merge(transaction,count_trans,on='msno')
    
    transaction = pd.merge(transaction,sum_trans,on='msno')
    
    return transaction 
    
    
```

```python
transaction_cv = transaction[transaction['transaction_date'].ge(20161232)]
```

```python
transaction_m_cv = featurization_transac(transaction_cv,member)
```

```python
    
def featurization_userlog(user_logs):



    
    sum_25 = user_logs.groupby(['msno'])['num_25'].sum()
    sum_50 = user_logs.groupby(['msno'])['num_50'].sum()
    sum_75 = user_logs.groupby(['msno'])['num_75'].sum()
    sum_985 = user_logs.groupby(['msno'])['num_985'].sum()
    sum_100 = user_logs.groupby(['msno'])['num_100'].sum()
    sum_unq = user_logs.groupby(['msno'])['num_unq'].sum()
    sum_sec = user_logs.groupby(['msno'])['total_secs'].sum()
    
    user_log_num_count = user_logs.groupby(['msno'])['num_25'].count()

    
    std_25 = user_logs.groupby(['msno'])['num_25'].std()
    std_50 =user_logs.groupby(['msno'])['num_50'].std()
    std_75 = user_logs.groupby(['msno'])['num_75'].std()
    std_985 = user_logs.groupby(['msno'])['num_985'].std()
    std_100 =user_logs.groupby(['msno'])['num_100'].std()
    std_unq = user_logs.groupby(['msno'])['num_unq'].std()
    std_sec = user_logs.groupby(['msno'])['total_secs'].std()
    
    user_log_num25_avg = sum_25/user_log_num_count 
    user_log_num50_avg = sum_50/user_log_num_count 
    user_log_num75_avg = sum_75/user_log_num_count 
    user_log_num985_avg = sum_985/user_log_num_count 
    user_log_num100_avg = sum_100/user_log_num_count 
    user_log_num_unq_avg = sum_unq/user_log_num_count 
    user_log_total_sec_avg = sum_sec/user_log_num_count 
    
    
    user_log_num25_avg = user_log_num25_avg.to_frame()
    user_log_num25_avg = user_log_num25_avg.rename(columns={'num_25':'num_25_avg'})
    user_log_num50_avg = user_log_num50_avg.to_frame()
    user_log_num50_avg = user_log_num50_avg.rename(columns={'num_50':'num_50_avg'})
    user_log_num75_avg = user_log_num75_avg.to_frame()
    user_log_num75_avg = user_log_num75_avg.rename(columns={'num_75':'num_75_avg'})
    user_log_num985_avg = user_log_num985_avg.to_frame()
    user_log_num985_avg = user_log_num985_avg.rename(columns={'num_985':'num_985_avg'})
    user_log_num100_avg = user_log_num100_avg.to_frame()
    user_log_num100_avg = user_log_num100_avg.rename(columns={'num_100':'num_100_avg'})
    user_log_num_unq_avg = user_log_num_unq_avg.to_frame()
    user_log_num_unq_avg = user_log_num_unq_avg.rename(columns={'num_unq':'num_unq_avg'})
    user_log_total_sec_avg = user_log_total_sec_avg.to_frame()
    user_log_total_sec_avg = user_log_total_sec_avg.rename(columns={'total_sec':'total_sec_avg'})
    

    final_data = pd.merge(sum_25,sum_50,on='msno')
    final_data = pd.merge(final_data,sum_75,on='msno')
    final_data = pd.merge(final_data,sum_985,on='msno')
    final_data = pd.merge(final_data,sum_100,on='msno')
    final_data = pd.merge(final_data,sum_unq,on='msno')
    final_data = pd.merge(final_data,sum_sec,on='msno')
    
    final_data = pd.merge(final_data,std_25,on='msno')
    final_data = pd.merge(final_data,std_50,on='msno')
    final_data = pd.merge(final_data,std_75,on='msno')
    final_data = pd.merge(final_data,std_985,on='msno')
    final_data = pd.merge(final_data,std_100,on='msno')
    final_data = pd.merge(final_data,std_unq,on='msno')
    final_data = pd.merge(final_data,std_sec,on='msno')
    
    final_data = pd.merge(final_data,user_log_num25_avg,on='msno')
    final_data = pd.merge(final_data,user_log_num50_avg,on='msno')
    final_data = pd.merge(final_data,user_log_num75_avg,on='msno')
    final_data = pd.merge(final_data,user_log_num985_avg,on='msno')
    final_data = pd.merge(final_data,user_log_num100_avg,on='msno')
    final_data = pd.merge(final_data,user_log_num_unq_avg,on='msno')
    final_data = pd.merge(final_data,user_log_total_sec_avg,on='msno')
    
    final_data = pd.merge(final_data,user_log_num_count,on='msno')
    
    final_data = pd.merge(final_data,user_logs,on='msno')
    
    

    
    return final_data
    
    
```

```python
## Load cv and test data 
FINAL_TRAIN = pd.read_csv('final_train_data.csv')
FINAL_CV = pd.read_csv('final_cv_data.csv')
FINAL_TEST = pd.read_csv('final_test_data.csv')
```

```python
FINAL_CV = FINAL_CV.dropna()
FINAL_TEST = FINAL_TEST.dropna()
```

```python
##### Load id's 

train_id = pd.read_csv('train.csv')
test_id = pd.read_csv('churn_comp_refresh/train_v2.csv')
```

```python
### Combine id's 

FINAL_TRAIN = pd.merge(FINAL_TRAIN,train_id,on='msno')
FINAL_CV = pd.merge(FINAL_CV,train_id,on='msno')
FINAL_TEST = pd.merge(FINAL_TEST,test_id,on='msno')
```

```python
print(FINAL_TRAIN.shape,FINAL_CV.shape,FINAL_TEST.shape)
```

```python
FINAL_TRAIN= FINAL_TRAIN.drop('Unnamed: 0',axis=1)
FINAL_CV = FINAL_CV.drop('Unnamed: 0',axis=1)
FINAL_TEST = FINAL_TEST.drop('Unnamed: 0',axis=1)
```

```python
print(FINAL_TRAIN.shape,FINAL_CV.shape,FINAL_TEST.shape)
```

```python
FINAL_TRAIN = FINAL_TRAIN.drop(['msno','gender'],axis=1)
FINAL_CV = FINAL_CV.drop(['msno','gender'],axis=1)
FINAL_TEST = FINAL_TEST.drop(['msno','gender'],axis=1)
```

```python
def normalize(data):
    for column in data.columns:
        max_val = data[column].max()
        min_val = data[column].min()
        deno = max_val-min_val
        if deno!=0:
            data[column]= (data[column]-min_val/max_val-min_val)
    return data 

FINAL_TRAIN = normalize(FINAL_TRAIN) 
FINAL_CV = normalize(FINAL_CV)
FINAL_TEST = normalize(FINAL_TEST)
```

```python
FINAL_TRAIN
```

```python
### X_train,X_test
X_train = FINAL_TRAIN.drop('is_churn',axis=1).values
Y_train = FINAL_TRAIN['is_churn'].values
print(X_train.shape,Y_train.shape)

### X _cv and Y_cv
X_cv = FINAL_CV.drop('is_churn',axis=1).values
Y_cv = FINAL_CV['is_churn'].values
print(X_cv.shape,Y_cv.shape)

### X_test and Y_test
X_test = FINAL_TEST.drop('is_churn',axis=1).values
Y_test = FINAL_TEST['is_churn'].values
print(X_test.shape,Y_test.shape)
```

```python
from sklearn import preprocessing
scaler  = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_cv = scaler.transform(X_cv)
```
