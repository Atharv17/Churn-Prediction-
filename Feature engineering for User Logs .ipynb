# Feature from User_Logs Table 

```python
import pandas as pd
```

```python
## Loading 10M rows 
user_logs = pd.read_csv("user_logs.csv",nrows=10000000)
```

```python
user_logs
```

```python
train = pd.read_csv("train.csv")
```

```python
train
```

```python
user_log_train = pd.merge(user_logs,train,on='msno',how='left')
```

```python
user_log_train
```

```python
user_log_before_jan = user_log_train[user_log_train['date'].le(20161231)]
```

```python
user_log_before_jan
```

```python
## Sum 
user_log_num25_sum = user_log_before_jan.groupby(['msno'])['num_25'].sum()
user_log_num50_sum = user_log_before_jan.groupby(['msno'])['num_50'].sum()
user_log_num75_sum = user_log_before_jan.groupby(['msno'])['num_75'].sum()
user_log_num985_sum = user_log_before_jan.groupby(['msno'])['num_985'].sum()
user_log_num100_sum = user_log_before_jan.groupby(['msno'])['num_100'].sum()
user_log_num_unq_sum = user_log_before_jan.groupby(['msno'])['num_unq'].sum()
user_log_total_sec_sum = user_log_before_jan.groupby(['msno'])['total_secs'].sum()
```

```python
## Saving all the values to pickle file 
user_log_num25_sum.to_pickle('user_log_num25_sum.pkl')
user_log_num50_sum.to_pickle('user_log_num50_sum.pkl')
user_log_num75_sum.to_pickle('user_log_num75_sum.pkl')
user_log_num985_sum.to_pickle('user_log_num985_sum.pkl')
user_log_num100_sum.to_pickle('user_log_num100_sum.pkl')
user_log_num_unq_sum.to_pickle('user_log_num_unq_sum.pkl')
user_log_total_sec_sum.to_pickle('user_log_total_sec_sum.pkl')
```

```python
# Count will be same for num_25,num_75 etc.
user_log_num_count = user_log_before_jan.groupby(['msno'])['num_25'].count()
user_log_num_count.to_pickle('user_log_num_count.pkl')
```

```python
## Calculating the Standard Deviation 
user_log_num25_std = user_log_before_jan.groupby(['msno'])['num_25'].std()
user_log_num50_std = user_log_before_jan.groupby(['msno'])['num_50'].std()
user_log_num75_std = user_log_before_jan.groupby(['msno'])['num_75'].std()
user_log_num985_std = user_log_before_jan.groupby(['msno'])['num_985'].std()
user_log_num100_std = user_log_before_jan.groupby(['msno'])['num_100'].std()
user_log_num_unq_std = user_log_before_jan.groupby(['msno'])['num_unq'].std()
user_log_total_sec_std = user_log_before_jan.groupby(['msno'])['total_secs'].std()
```

```python
## Saving all the values in Pickle file 
user_log_num25_std.to_pickle('user_log_num25_std.pkl')
user_log_num50_std.to_pickle('user_log_num50_std.pkl')
user_log_num75_std.to_pickle('user_log_num75_std.pkl')
user_log_num985_std.to_pickle('user_log_num985_std.pkl')
user_log_num100_std.to_pickle('user_log_num100_std.pkl')
user_log_num_unq_std.to_pickle('user_log_num_unq_std.pkl')
user_log_total_sec_std.to_pickle('user_log_total_sec_std.pkl')
```

```python
## Calculating the average by using sum and count 
user_log_num25_avg = user_log_num25_sum/user_log_num_count 
user_log_num50_avg = user_log_num50_sum/user_log_num_count 
user_log_num75_avg = user_log_num75_sum/user_log_num_count 
user_log_num985_avg = user_log_num985_sum/user_log_num_count 
user_log_num100_avg = user_log_num100_sum/user_log_num_count 
user_log_num_unq_avg = user_log_num_unq_sum/user_log_num_count 
user_log_total_sec_avg = user_log_total_sec_sum/user_log_num_count 
```

```python
user_log_num25_avg = user_log_num25_avg.to_frame()
user_log_num25_avg = user_log_num25_avg.rename(columns={'num_25':'num_25_avg'})
user_log_num50_avg = user_log_num50_avg.to_frame()
user_log_num50_avg = user_log_num50_avg.rename(columns={'num_50':'num_50_avg'})
user_log_num75_avg = user_log_num75_avg.to_frame()
user_log_num75_avg = user_log_num75_avg.rename(columns={'num_75':'num_75_avg'})
user_log_num985_avg = user_log_num985_avg.to_frame()
user_log_num985_avg = user_log_num985_avg.rename(columns={'num_985':'num_985_avg'})
user_log_num100_avg = user_log_num100_avg.to_frame()
user_log_num100_avg = user_log_num100_avg.rename(columns={'num_100':'num_100_avg'})
user_log_num_unq_avg = user_log_num_unq_avg.to_frame()
user_log_num_unq_avg = user_log_num_unq_avg.rename(columns={'num_unq':'num_unq_avg'})
user_log_total_sec_avg = user_log_total_sec_avg.to_frame()
user_log_total_sec_avg = user_log_total_sec_avg.rename(columns={'total_sec':'total_sec_avg'})
```

```python
## Saving all the average values to the pickle file 
user_log_num25_avg.to_pickle('user_log_num25_avg.pkl') 
user_log_num50_avg.to_pickle('user_log_num50_avg.pkl')
user_log_num75_avg.to_pickle('user_log_num75_avg.pkl')
user_log_num985_avg.to_pickle('user_log_num985_avg.pkl')  
user_log_num100_avg.to_pickle('user_log_num100_avg.pkl')
user_log_num_unq_avg.to_pickle('user_log_num_unq_avg.pkl')
user_log_total_sec_avg.to_pickle('user_log_total_sec_avg.pkl')
```

```python
final_data_user_logs = pd.merge(user_log_num25_sum,user_log_num25_std,on='msno')
```

```python
final_data_user_logs = pd.merge(final_data_user_logs,user_log_num25_avg,on='msno')
final_data_user_logs = pd.merge(final_data_user_logs,user_log_num_count,on='msno')
```

```python


final_data_user_logs = pd.merge(final_data_user_logs,user_log_num75_sum,on='msno')
final_data_user_logs = pd.merge(final_data_user_logs,user_log_num75_std,on='msno')
```

```python

final_data_user_logs = pd.merge(final_data_user_logs,user_log_num985_sum,on='msno')
final_data_user_logs = pd.merge(final_data_user_logs,user_log_num985_std,on='msno')
```

```python

final_data_user_logs = pd.merge(final_data_user_logs,user_log_num_unq_sum,on='msno')
final_data_user_logs = pd.merge(final_data_user_logs,user_log_num_unq_std,on='msno')
```

```python

final_data_user_logs = pd.merge(final_data_user_logs,user_log_total_sec_sum,on='msno')
final_data_user_logs = pd.merge(final_data_user_logs,user_log_total_sec_std,on='msno')
```

```python
final_data_user_logs = pd.merge(final_data_user_logs,user_log_num_count,on='msno')
```

```python
final_data_user_logs = pd.merge(final_data_user_logs,user_logs,on='msno')
```

```python
final_data_user_logs = final_data_user_logs.dropna()
```
